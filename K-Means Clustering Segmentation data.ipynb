```python
df.drop(['ID'],inplace=True,axis=1)
```

```python
df.head()
```

```python
X=df.drop(['Settlement size'],axis=1)
Y=df['Settlement size']
```

```python
from sklearn.preprocessing import MinMaxScaler
mns=MinMaxScaler()
X=mns.fit_transform(X)
cols=['Sex','Martial status','Age','Education','Income','Occupation']
X=pd.DataFrame(X,columns=cols)
```

```python
X.head()
```

```python
# Model building with two clusters
```

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=1,random_state=42)
kmeans.fit(X)
```

```python
kmeans.cluster_centers_
```

```python
kmeans.inertia_
```

```python
labels=kmeans.labels_
```

```python
correct_labels=sum(Y==labels)
```

```python
print("results are :",correct_labels,"out of",Y.size)
```

```python
print("Correctly labelled are: ",(correct_labels/Y.size))
```
